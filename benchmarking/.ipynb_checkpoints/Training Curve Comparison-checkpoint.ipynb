```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from micrograd.engine import Value
from micrograd.nn import MLP
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
# --- Data ---
X, y = make_moons(n_samples=200, noise=0.2, random_state=42)
y = y * 2 - 1  # make it -1 or 1
X_torch = torch.tensor(X, dtype=torch.float32)
y_torch = torch.tensor(y, dtype=torch.float32).view(-1, 1)
```

```python
# --- Micrograd Model ---
model_micro = MLP(2, [16, 16, 1])
micro_loss_list = []
micro_acc_list = []
```

```python
def micro_loss(X, y):
    inputs = [list(map(Value, xrow)) for xrow in X]
    scores = list(map(model_micro, inputs))
    losses = [(1 + -yi * score).relu() for yi, score in zip(y, scores)]
    data_loss = sum(losses) * (1.0 / len(losses))
    reg_loss = 1e-4 * sum((p * p for p in model_micro.parameters()))
    total_loss = data_loss + reg_loss
    accuracy = sum([(yi > 0) == (score.data > 0) for yi, score in zip(y, scores)]) / len(y)
    return total_loss, accuracy
```

```python
for epoch in range(100):
    loss, acc = micro_loss(X, y)
    model_micro.zero_grad()
    loss.backward()
    lr = 1.0 - 0.9 * epoch / 100
    for p in model_micro.parameters():
        p.data -= lr * p.grad
    micro_loss_list.append(loss.data)
    micro_acc_list.append(acc * 100)
```

```python
# --- PyTorch Model ---
class TorchMLP(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(2, 16)
        self.fc2 = nn.Linear(16, 16)
        self.fc3 = nn.Linear(16, 1)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        return self.fc3(x)
```

```python
model_torch = TorchMLP()
optimizer = torch.optim.SGD(model_torch.parameters(), lr=1.0)
torch_loss_list = []
torch_acc_list = []
```

```python
for epoch in range(100):
    model_torch.train()
    optimizer.zero_grad()
    scores = model_torch(X_torch)
    losses = torch.relu(1 + -y_torch * scores)
    data_loss = losses.mean()
    reg_loss = 1e-4 * sum((p ** 2).sum() for p in model_torch.parameters())
    loss = data_loss + reg_loss
    loss.backward()
    lr = 1.0 - 0.9 * epoch / 100
    for param_group in optimizer.param_groups:
        param_group['lr'] = lr
    optimizer.step()
    
    preds = (scores > 0).float()
    acc = (preds == (y_torch > 0).float()).float().mean().item()
    
    torch_loss_list.append(loss.item())
    torch_acc_list.append(acc * 100)
```

```python
# --- Plot ---
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.plot(micro_loss_list, label="Micrograd")
plt.plot(torch_loss_list, label="PyTorch")
plt.title("Loss vs Epochs")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(micro_acc_list, label="Micrograd")
plt.plot(torch_acc_list, label="PyTorch")
plt.title("Accuracy vs Epochs")
plt.xlabel("Epochs")
plt.ylabel("Accuracy (%)")
plt.legend()

plt.tight_layout()
plt.show()
```

```python
def summarize_training(loss_list, acc_list, last_n=10):
    final_loss = loss_list[-1]
    final_acc = acc_list[-1]
    
    best_acc = max(acc_list)
    best_epoch = acc_list.index(best_acc) + 1  # +1 for 1-based epoch count
    
    avg_loss_last_n = sum(loss_list[-last_n:]) / last_n
    avg_acc_last_n = sum(acc_list[-last_n:]) / last_n
    
    loss_improvement = loss_list[0] - final_loss
    acc_improvement = final_acc - acc_list[0]
    
    # Epoch to reach 90% of best accuracy (if possible)
    threshold = 0.9 * best_acc
    epochs_to_threshold = next((i+1 for i, acc in enumerate(acc_list) if acc >= threshold), None)
    
    return {
        "final_loss": final_loss,
        "final_accuracy": final_acc,
        "best_accuracy": best_acc,
        "best_epoch": best_epoch,
        "avg_loss_last_{}": avg_loss_last_n,
        "avg_accuracy_last_{}": avg_acc_last_n,
        "loss_improvement": loss_improvement,
        "accuracy_improvement": acc_improvement,
        "epochs_to_90%_best_accuracy": epochs_to_threshold
    }

# Summarize micrograd
micro_stats = summarize_training(micro_loss_list, micro_acc_list)

# Summarize pytorch
torch_stats = summarize_training(torch_loss_list, torch_acc_list)

print("Micrograd Training Summary:")
for k, v in micro_stats.items():
    print(f"{k}: {v}")

print("\nPyTorch Training Summary:")
for k, v in torch_stats.items():
    print(f"{k}: {v}")
```
